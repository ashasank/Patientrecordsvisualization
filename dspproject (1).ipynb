```python
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['daily_internet_use'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Daily Internet Use')
plt.ylabel('Count')
plt.title('Daily Internet Use Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['available_vehicles'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Available Vehicles')
plt.ylabel('')
plt.show()
```

```python
df['dob'] = pd.to_datetime(df['dob'])
df['year'] = df['dob'].dt.year
yearly_counts = df['year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Yearly Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='employment_status', y='avg_commute', data=df)
plt.title('Average Commute Time by Employment Status')
plt.ylabel('Average Commute Time')
plt.xlabel('Employment Status')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='daily_internet_use', y='avg_commute', data=df)
plt.title('Daily Internet Use vs. Average Commute Time')
plt.xlabel('Daily Internet Use')
plt.ylabel('Average Commute Time')
plt.show()
```

```python
# Selecting numeric columns for correlation matrix
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

```python
employment_status_counts = df.groupby(['employment_status', 'gender']).size().unstack()
employment_status_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Gender Distribution by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='education', y='avg_commute', hue='employment_status', data=df)
plt.title('Average Commute Time by Education Level and Employment Status')
plt.xlabel('Education Level')
plt.ylabel('Average Commute Time')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='gender', y='daily_internet_use', data=df)
plt.title('Daily Internet Use Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Daily Internet Use')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your data is in a DataFrame named 'df'
# df = pd.read_csv('your_dataset.csv')

# Count the occurrences of each disease
disease_counts = df['disease'].value_counts()

# Bar Chart for Diseases
plt.figure(figsize=(10, 6))
disease_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.title('Different Types of Diseases')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import pandas as pd

# Assuming your data is in a DataFrame named 'df'
# df = pd.read_csv('your_dataset.csv')

# Extract unique values from the 'disease' column
unique_diseases = df['disease'].unique()

# Display unique diseases
print("Different types of diseases present in the dataset:")
for disease in unique_diseases:
    print(disease)
```

```python
grouped_data = df.groupby(['ancestry', 'disease'])['children'].sum().reset_index()

# Display the results
print("Children affected by different diseases in different countries:")
print(grouped_data)
```

```python
plt.figure(figsize=(10, 6))
for ancestry in grouped_data['ancestry'].unique():
    ancestry_data = grouped_data[grouped_data['ancestry'] == ancestry]
    plt.bar(ancestry_data['disease'], ancestry_data['children'], label=ancestry)

plt.xlabel('Disease')
plt.ylabel('Number of Children Affected')
plt.title('Children Affected by Different Diseases in Different Ancestries')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()




```

```python
import folium
map_center = [df['zipcode'].mean(), df['zipcode'].mean()]  # Center the map based on the average of zip codes
map_zoom = 6
mymap = folium.Map(location=map_center, zoom_start=map_zoom)

# Add markers for military service 'Yes'
for index, row in df.iterrows():
    if row['military_service'] == 'Yes':
        folium.Marker(
            location=[row['zipcode'], row['zipcode']],
            popup=f"ID: {row['id']}<br>Military Service: {row['military_service']}",
            icon=folium.Icon(color='green')
        ).add_to(mymap)
```

```python
mymap.save('military_service_map.html')
```
